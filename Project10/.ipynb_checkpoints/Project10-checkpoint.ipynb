# Загрузка и подготовка данных

```python
import time
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso, LinearRegression
from lightgbm import LGBMRegressor
import warnings
warnings.simplefilter("ignore")
```

## Загрузка данных

```python
df = pd.read_csv('/datasets/autos.csv')
df
```

## Изучение и обработка данных

```python
df.info()
```

Пропуски имеются в категориальных признаках, в количественных их нет


Признаки DateCrawled, DateCreated, LastSeen, NumberOfPictures, postalcode, lastseen не несут никакой ценности, поэтому дропнем их

```python
df = df.dropna()
```

```python
df.info()
```

```python
df = df.drop(['DateCrawled', 'RegistrationMonth', 
              'DateCreated', 'PostalCode', 'LastSeen', 'NumberOfPictures'], axis=1)
```

```python
corr = df.corr()
plt.figure(figsize = (10,10))
sb.heatmap(corr,annot = True,linewidths=.5,cmap="YlGnBu")
```

Количественные признаки коррелирующие с ценой:
- год регистрации
- мощность
- пробег


Проверим дата сет на дубликаты

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.duplicated().sum()
```

Дубликатов больше нет

## Подготовка данных

```python
columns = ['VehicleType','Gearbox','Model','FuelType','Brand','NotRepaired']
df[columns] = df[columns].astype('category')
enc = OrdinalEncoder()
for col in columns:
    df[columns] = enc.fit_transform(df[columns])
```

```python
X = df.drop('Price', axis=1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=12345)

print(f'X train/test: {X_train.shape, X_test.shape}')
print(f'y train/test: {y_train.shape, y_test.shape}')
```

```python
X_train
```

# Обучение моделей

```python
def rmse(y_train, prediction):
    return mean_squared_error(y_train, prediction) ** .5

rmse_score = make_scorer(rmse, greater_is_better=False)
```

## Lasso

```python
lasso = Lasso(max_iter=100, random_state=12345)
parametrs_lasso= {'alpha': range(-2, 5, 1)}
gs_lasso = GridSearchCV(lasso, parametrs_lasso, cv=5, scoring=rmse_score)
```

```python
%%time
gs_lasso.fit(X_train, y_train)
```

```python
gs_lasso.best_params_
```

```python
gs_lasso.best_score_
```

## LightGBM Regressor

```python
lgbm = LGBMRegressor(boosting_type='gbdt', random_state=12345)
parametrs_lgbm = {'n_estimators': range(50, 150, 50),
                  'num_leaves': range(5, 40, 10), 
                  'learning_rate': range(-2, 5, 1)}
gs_lgbm = GridSearchCV(lgbm, parametrs_lgbm, cv=5, scoring=rmse_score)
```

```python
%%time
gs_lgbm.fit(X_train, y_train)
```

```python
gs_lgbm.best_params_
```

```python
gs_lgbm.best_score_
```

## CatBoostRegressor

```python
cat_boost = CatBoostRegressor(loss_function='RMSE', random_state=12345)
parametrs_cb = {'n_estimators': range(50, 150, 50)}
gs_cb = GridSearchCV(cat_boost, parametrs_cb, cv=5, scoring=rmse_score)
```

```python
%%time
gs_cb.fit(X_train, y_train)
```

```python
gs_cb.best_params_
```

```python
gs_cb.best_score_
```

# Анализ моделей

```python
models = [
    gs_lasso.best_estimator_, 
    gs_lgbm.best_estimator_, 
    gs_cb.best_estimator_
]

for model in models:
    start = time.time()
    model.fit(X_train, y_train)
    fit_time = time.time() - start
    pred = model.predict(X_test)
    pred_time = time.time() - start
    score = rmse(y_test, pred)
    print(f'Название модели: {model}\nВремя обучения: {fit_time} сек.\nВремя предсказания: {pred_time} сек. \nРезультат RMSE: {score}')
    print()
```
